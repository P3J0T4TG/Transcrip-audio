```python
import os
from pathlib import Path

# Carpeta con los audios
audio_dir = Path("/home/p3j0t4/Documentos/9.Samu")

# Obtener archivos de audio
audio_files = sorted(audio_dir.glob("*.opus"))
print(f"Archivos encontrados: {len(audio_files)}")
for f in audio_files:
    print(f"  - {f.name}")
```

```python
# Cargar modelo Whisper (opciones: tiny, base, small, medium, large-v2, large-v3)
# 'small' ofrece buen equilibrio entre velocidad y precisión para español
from faster_whisper import WhisperModel

model = WhisperModel("small", device="cpu", compute_type="int8")
print("Modelo cargado correctamente")
```

```python
# Transcribir cada archivo de audio
for audio_path in audio_files:
    print(f"\nTranscribiendo: {audio_path.name}")
    print("-" * 60)
    
    segments, info = model.transcribe(str(audio_path), language="es", beam_size=5)
    
    # Unir todos los segmentos en texto completo
    texto_completo = " ".join(segment.text.strip() for segment in segments)
    
    # Guardar transcripción en archivo .txt junto al audio
    output_path = audio_path.with_suffix(".txt")
    output_path.write_text(texto_completo, encoding="utf-8")
    
    print(f"Idioma detectado: {info.language} (prob: {info.language_probability:.2f})")
    print(f"Guardado en: {output_path.name}")
    print(f"Fragmento: {texto_completo[:300]}...")

print("\n✓ Todas las transcripciones completadas")
```
